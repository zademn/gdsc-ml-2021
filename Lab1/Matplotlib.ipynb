```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(9)
y = np.arange(10, 19)
```

```python
plt.figure(figsize = (7, 7))
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize = (7, 7))
plt.scatter(x, y)
plt.show()
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
y2 = np.cos(x)

# plotting
plt.plot(x, y, c ='red', label = 'sin')
plt.plot(x, y2, c = 'blue', label = 'cos')
plt.xlabel("sin range")
plt.ylabel("cos range")
plt.legend()
plt.show()
```

```python
# mx + n
n = 10
real_w = 2
real_b = .5
x = [np.random.random() for _ in range(n)]  # Sample n points between 0, 1
y = [real_w * xi + real_b + .1 * np.random.random()
     for xi in x] 
```

```python
plt.figure(figsize = (7, 7))
plt.scatter(x, y, s = 100, marker = 'x')
plt.show()
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
y2 = np.cos(x)

# plotting
plt.figure(figsize = (10, 10))
# plot on first plot
plt.subplot(2, 2, 1)
plt.plot(x, y, c ='red', label = 'sin')
plt.xlabel("sin range")

plt.subplot(2, 2, 2)
plt.plot(x, y2, c = 'blue', label = 'cos')


plt.subplot(2, 2, 3)
plt.plot(x, x**2)

plt.subplot(2, 2, 4)
plt.plot(x, np.sqrt(x))



# plt.legend()
plt.show()
```
